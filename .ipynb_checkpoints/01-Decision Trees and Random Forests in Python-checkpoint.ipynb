
# Les arbres de décisions et Random Forest en Python

Cette partie est une application des arbres de décisions et de Random Forest en Python. le dataset est un dataset des patients souffrant de la cyphose (kyphosis en anglais) qui est une maladie qui déforme la colonne vertébrale. Dans Le dataset, chaque ligne représente un enfant qui a eu une chirurgie corrective de la colonne vertébrale pour corriger la cyphose. Age représente l'age en mois, number: le nombre de vertèbres touchées, Start: l'indice de la première vertèbre qui a été opérée en partant du haut de la colonne vertébrale. Donc le but c'est de prédire si une opération va être fructueuse en fonction des 3 derniers features.

## Importation de librairies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Charger les données  

```python
df = pd.read_csv('kyphosis.csv')
```

```python
df.head()
```

## Analyse explorative des données 
Nous allons utilisons la méthode pairplot pour ce petit DataSet.


```python
sns.pairplot(df,hue='Kyphosis',palette='Set1')
```

## Séparation de l'ensemble d'apprentissage et ensemble de test
Divisons maintenant les données en données d'entrainement et données de test


```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
```

## l'arbre de décision
Nous allons commencer par entrainer un seul arbre de décision.



```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## La prédiction et l'évaluation

Nous allons maintenant évaluer notre arbre de décision

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
###################################
for i in range(20,100):    
    t_s = i/100
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=t_s, random_state=101)
    dtree = DecisionTreeClassifier()
    dtree.fit(X_train ,y_train)
    predictions = dtree.predict(X_test)
    print("test_size={} -> {}".format(t_s ,classification_report(y_test,predictions)[179:183]))
###################################
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

#import graphviz

Image(graph.create_png())
```

## Random Forest

Comparons maintenant l'arbre de décision à la méthode Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
prec = []
for i in range(1,201):
    rfc = RandomForestClassifier(n_estimators=i)
    rfc.fit(X_train, y_train)

    rfc_pred = rfc.predict(X_test)

    prec.append( float(classification_report(y_test,rfc_pred)[179:183]) )
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,201),prec,color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=8)
plt.title("Precision en fonction de la valeur de n_estimators")
plt.xlabel('K')
plt.ylabel("Taux d'erreur")
```
